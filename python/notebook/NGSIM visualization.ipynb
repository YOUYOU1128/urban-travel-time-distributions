Get the data
--------------------

- Download the NGSIM dataset [here](http://ngsim-community.org/)
    - You will need to create an account
    - This work focuses on the Peachtree dataset (4pm-4:15pm)
- Set the environment variable <code> $MM_DATA_DIR </code> as the root of the folder containing the NGSIM data


```python
'''
Created on Mar 8, 2013

@author: audehofleitner
'''

import pandas as pd
from pandas import DataFrame, Series
from datetime import datetime, timedelta
import os

FIELDS = (['veh_id', 'frame_id', 'epoch_time', 'local_x', 'local_y',
           'veh_vel', 'lane_id', 'o_zone', 'd_zone', 'intersection',
           'section', 'direction', 'mvmt'])

def get_data(direction, network='Peachtree', time='0400pm-0415pm'):
  """
  direction: 2 - north-bound (NB) or 4 - south-bound (SB)
  network: Peachtree
  time: 0400pm-0415pm
  """
  fname = '{0}/NGSIM/{1}/vehicle-trajectory-data/{2}/trajectories-{2}.csv'.format(data_dir(), network, time)
  data = pd.read_csv(fname)
  data = data[FIELDS]
  data = data[data['mvmt'] == 1]
  data = data[data['direction'] == direction]
  data['time'] = (data['epoch_time'] - data['epoch_time'].min()) / 1000.0
  data['date'] = data['time'].map(lambda d: datetime(2008, 11, 8, 16, 0, 0) + timedelta(seconds=d))
  data['local_y'] = data['local_y'] / 0.3048
  data['local_x'] = data['local_x'] / 0.3048
  data['veh_vel'] = data['veh_vel'] * 1.09728
  return data

def get_travel_times(df):
  df = df[df['section'] != 0]
  g = df['time'].groupby([df['veh_id'], df['section']])
  res = DataFrame(g.max() - g.min())
  res.columns = ['tt']
  res['time'] = g.min()
  res = res.reset_index()
  return res

def data_dir():
  """ Directory that contains the local data.

  The default value is overriden if the environment variable
  MM_DATA_DIR is defined. In this case, the value of this variable
  is returned instead.
  """
  if 'MM_DATA_DIR' in os.environ:
    return os.environ['MM_DATA_DIR']
  raise Exception('Check that the environment variable MM_DATA_DIR is setup')

```

### Trajectories

```python
def plot_trajectories(data):
  x = data['time']
  y = data['local_y']
  z = data['veh_vel']
  cm = matplotlib.cm.get_cmap('bone')
  p = plt.scatter(x, y, c=z, marker='.', edgecolor='none', s=1, cmap=cm)
  cb = plt.colorbar(p)
  cb.set_label('Speed (km/h)', fontsize=14)
  plt.xlabel('Time (s)')
  plt.ylabel('Offset (m)')
```

```python
# Gets the data for Northbound direction (south bound is 4)
df = get_data(2)
```

```python
# Plot the trajectories
fig = plt.figure(figsize=(12,6))
plot_trajectories(df)
plt.xlim(0, 1000)
plt.ylim(0, 7000)
plt.savefig('{0}/{1}'.format(FIGURES_DIR, 'traj1.pdf'))
```

### Time series and distribution

```python
g = df['date'].groupby([df['veh_id'], df['section']])
res = DataFrame([(g.max() - g.min()), g.min()]).T
res.columns = ['tt', 'date']
res['tt'] = res['tt'].map(timedelta.total_seconds)
res = res.reset_index()
res = res.set_index('date')
res = res.sort()
tt = res
sections = [2, 3, 5]
```

```python
from sklearn import mixture
import numpy as np
N_mixt = 4

def learn_gmm(sec_tt):
    low_bic = np.infty
    best_gmm = None
    for n_mixt in range(1, N_mixt + 1):
        gmm = mixture.GMM(n_components=n_mixt)
        gmm.fit(sec_tt)
        if gmm.bic(sec_tt) < low_bic:
            low_bic = gmm.bic(sec_tt)
            best_gmm = gmm
    return best_gmm
```

```python
fig = plt.figure(figsize=(12,8))
cax = fig.add_axes([0.1, 0.02, 0.8, 0.04])

w = 0.3
h = 0.4
a = []
images = []
ttmin = 1e40
ttmax = -1e40

for i, s in enumerate(sections):
    sec_tt = tt[tt['section'] == s]
    pos = [0.05 + 1.1 * i * w, 0.08, w, h]
    ax = fig.add_axes(pos)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ts = sec_tt['tt'].resample('5S', how='median').fillna(0)
    resamp_tt = ts.values
    ttmin = min(np.min(resamp_tt), ttmin)
    ttmax = max(np.max(resamp_tt), ttmax)
    valinit = 20

    temp_tt = np.zeros(valinit * (1 + np.floor(len(resamp_tt) / valinit)))
    temp_tt[: len(resamp_tt)] = resamp_tt
    images.append(ax.pcolormesh(temp_tt.reshape((-1, valinit)), cmap=cm.bone_r))
    a.append(ax)
    
    # sec_tt['tt'].plot(style='-+')
    plt.title('Segment {0}'.format(s), fontsize=14)
   # if i == 0:
  #      plt.ylabel('Travel time (s)', fontsize=14)
 #   if i == len(sections) / 2:
#        plt.xlabel('Time (s)', fontsize=14)
    
    pos = [0.05 + 1.1 * i * w, 0.08 + 1.25 * h, w, h]
    ax = fig.add_axes(pos)
    histo = ax.hist(sec_tt['tt'], 30, normed=1, color='b', alpha=.3, label='Histogram')
    gmm = learn_gmm(sec_tt['tt'])
    x = np.linspace(0, max(sec_tt['tt']) + 10, 500)
    ax.plot(x, np.exp(gmm.score(x)), '-b', linewidth=2, label='GMM fit')
    ax.text(0.6, 0.9, 'GMM fit:', transform = ax.transAxes)
    ax.text(0.6, 0.8, '{0} components'.format(gmm.n_components), transform = ax.transAxes)
    if i == len(sections) / 2:
        plt.xlabel('Travel time (s)', fontsize=14)
    if i == 0:
        plt.legend(loc=7)
        plt.ylabel('Probability distribution', fontsize=14)

class ImageFollower:
    'update image in response to changes in clim or cmap on another image'
    def __init__(self, follower):
        self.follower = follower
    def __call__(self, leader):
        self.follower.set_cmap(leader.get_cmap())
        self.follower.set_clim(leader.get_clim())

norm = colors.Normalize(vmin=ttmin, vmax=ttmax)
for i, im in enumerate(images):
    im.set_norm(norm)
    if i > 0:
        images[0].callbacksSM.connect('changed', ImageFollower(im))

# The colorbar is also based on this master image.
cbar = fig.colorbar(images[0], cax, orientation='horizontal')
cbar.set_label('Travel time (s)', fontsize=14)
plt.savefig('{0}/{1}'.format(FIGURES_DIR, 'tt_ts.pdf'), bbox_inches='tight')
```
